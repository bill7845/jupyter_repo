안녕하세요! [크몽](https://kmong.com) 데이터팀에 오신 것을 환영합니다.

크몽은 "비합리적인 서비스 시장을 정보기술로 혁신하여 새로운 일자리를 만들고 사람들이 행복하게 일하는 세상을 만든다."는 미션을 가지고 있는 프리랜서 마켓 플랫폼입니다. 크몽팀은 디자인, 마케팅, IT&개발, 컨텐츠 제작, 통번역 등 다양한 서비스 카테고리를 기반으로 직업과 직장의 경계를 허물며 사람들이 스스로 경제 주체가 되어 행복하게 일하는 세상을 만들고 있습니다.

크몽에서는 주어진 데이터를 활용하여 1) 사용자에게 새로운 상품을 추천하거나, 2) 오픈마켓에서 금기시되는 개인간 거래를 발견하 등 다양한 분야에서 데이터와 머신러닝 알고리즘을 사용합니다. 여기에는 크몽의 서비스에서 발생한 다양한 데이터를 사용하게 되는데, 이 부분에서 가장 중요한 것이 바로 데이터를 정리하는 과정, 일명 데이터 클리닝([Data Cleaning](https://en.wikipedia.org/wiki/Data_cleansing))입니다.

실제 데이터분석가로서 현업에 투입되면, 데이터를 분석하는 일보다 데이터를 정리하는 일에 더 많은 시간을 할애하게 됩니다. 통상적으로 전체 업무시간에 70% ~ 80%는 데이터를 정리하는 일을 하게 되는데, 이 과정에서 실력있는 데이터 분석가와 그렇지 않은 데이터 분석가의 차이가 눈에 보이게 됩니다. 실력있는 데이터 분석가는 동일한 분량의 데이터를 2~3시간 안에 처리하는데 비해, 실력이 부족하고 아직 툴에 익숙하지 않은 데이터 분석가는 적어도 반나절, 길면 하루 ~ 이틀 정도의 시간을 투입해야 겨우 동일한 분량의 데이터를 처리하곤 합니다.

그런 의미에서, 오늘 이 쥬피터 노트북을 받은 수강생분은 하루동안 크몽의 일일 데이터 컨설턴트로서 일을 할 것입니다. 여러분들은 오늘 크몽의 사용자 로그 데이터, 웹사이트/모바일 서비스에서 사용자들이 활동한 기록 등을 가져와, 이를 빠른 시간 안에 다듬고 정리하는 일을 맡게 될 것입니다. 하기에 제시한 15개의 데이터 분석 요청을 프로그래밍 언어 파이썬([Python](http://python.org/))과 파이썬의 데이터 분석 패키지 판다스([Pandas](https://www.python.org/))를 활용해 풀어주세요.

숙련된 데이터분석가의 경우 하기의 요청을 늦어도 2시간 내에는 해결할 수 있습니다. 즉, **2시간 안에 모든 문제를 풀 수 있다면 합격입니다.**

문제를 풀 때 다른 자료를 참고하거나, 구글에 검색하는 것 모두 허용합니다. **kmong-conversion.csv**

먼저 크몽의 로그 데이터를 하나씩 가져오겠습니다. 가장 처음 가져올 데이터는 크몽의 가장 핵심 데이터라고 할 수 있는 ```kmong-conversion.csv``` 입니다. 여기에는 사용자가 크몽의 웹사이트/모바일 서비스에서 활동한 모든 활동 기록(activity)이 상세하게 담겨져 있으며, 이 정보를 바탕으로 데이터를 분석하여 사용자의 패턴을 유추할 수 있습니다. 컬럼 정보는 다음과 같습니다.

* ```eventcategory```: 앱의 이벤트에 대한 카테고리입니다. 크게 다음의 종류가 있습니다.
  - ```install```: 앱설치
  - ```launch```: 앱실행 
  - ```deeplinkLaunch```: 딥링크 통한 앱실행
  - ```goal```: 앱 내에서의 활동 여부
  - ```exit```: 앱 종료
  - ```foreground```, background```: 앱을 foreground, background로 전환
  - ```launchlnSession```: Session의 시작.
* ```isfirstactivity```: 해당 이벤트가 해당 유저에 한하여 첫 번째인지 혹은 두 번째 이상인지를 알려주는 값입니다. 
  - boolean 형식으로, 해당 이벤트가 해당 유저에 한하여 첫 번째 이벤트이면 True, 아니면 False를 갖습니다.
* ```apppackagename```: 앱의 패키지의 고유한 이름입니다. 안드로이드의 경우 applicationId을, iOS의 경우 Bundle ID를 가져오게 됩니다.
* ```appversion```: kmong 어플리케이션의 버전입니다.
* ```devicetype```: 사용자가 kmong app을 실행하고 있는 기기의 제품명입니다.
* ```devicemanufacturer```: 사용자가 kmong app을 실행하고 있는 기기의 제조회사입니다. 
* ```osversion```: 사용자가 kmong app을 실행하고 있는 기기의 os 버전입니다.
* ```canonicaldeviceuuid```: 사용자가 kmong app을 실행하고 있는 기기의 고유 식별자입니다. (사용자를 구분하는 식별자로 이용할 수 있습니다)
* ```sourcetype```: 해당 로그를 만든 고객이 유입된 경로입니다. 유료 광고 채널, 바이럴 채널, 앱 마켓 등으로 구분합니다.
* ```channel```: sourcetype과 동일하지만 조금 더 세분화되어 있습니다.
  - unattributed, WEB, google-play, m_naver, google, (not set), google.adwords, m_naverpowercontents, pc_naver, apple.searchads, facebook, m_daum
* ```params_campaign```: 마케터가 입력한 캠페인 파라미터 중 캠페인명입니다. 
* ```params_medium```: 마케터가 입력한 캠페인 파라미터 중 매체입니다.
* ```params_term```: 마케터가 입력한 캠페인 파라미터 중 키워드입니다.
* ```inappeventcategory```: 인앱이벤트 식별자값으로 인앱이벤트를 지칭하는 3개의 값(category > action > label) 중 가장 상위에 속하는 구분값입니다.
  - ```eventcategory```가 goal인 경우에만 제공됩니다.
  - 차후에 읽어올 funnel 데이터셋과 연결할 수 있습니다.
  - ex) seller_selling_history.view, gig_detail.view
* ```inappeventlabel```: categoryid와 동일합니다.
  - 차후에 읽어올 category 데이터셋과 연결할 수 있습니다.
  - 인앱이벤트를 지칭하는 3개의 값(category > action > label) 중 마지막에 속하는 구분값입니다.
* ```eventdatetime```: 모바일 클라이언트 상에서의 실제 이벤트 발생시간입니다. ISO 8601 표준에 따라서 제공됩니다. 
* ```isfirstgoalactivity```: goal 이벤트의 경우, 해당 Goal Label, Description, Key, Category를 가진 이벤트가 해당 유저에 한하여 첫 번째인지 혹은 두 번째 이상 인지를 알려주는 boolean 값입니다. 예를 들어 동일한 유저로부터 두 번 이상의 같은 Goal 이벤트가 발생한 경우(e.g. 동일한 물건 재구매 등) 이 값은 false가 됩니다.

* ```event_rank```: 데이터를 정렬하는데 필요한 컬럼입니다. log엔 기록되지 않으며, DS_School에서 제공하는 전처리된 컬럼입니다.

이 데이터를 차후 데이터 분석용으로 다루기 쉽게, ```ostype(clean)```과 ```osversion(clean)```이라는 두 개의 컬럼으로 분리하고 싶습니다. 가령 에를 들자면

  * ```osversion``` 컬럼값이 iOS11.4.1이라면, 새로운 컬럼인 ```ostype(clean)```에는 iOS가, ```osversion(clean)```에는 11.4.1이 들어가야 합니다.
  * ```osversion``` 컬럼값이 Android7.0이라면, 새로운 컬럼인 ```ostype(clean)```에는 Android가, ```osversion(clean)```에는 7.0이 들어가야 합니다.

````python
# osversion 컬럼값을 활용해 OS의 타입(iOS / Android)을 가져오는
# find_ostype라는 이름의 함수를 정의합니다.
def find_ostype(osversion):
    # osversion에 iOS라는 값이 있으면 아이폰이라고 간주하고 iOS라는 값을 반환합니다.
    if "iOS" in osversion:
        return 'iOS'
    # 그렇지 않으면 안드로이드라고 간주하고 Android라는 값을 반환합니다.
    else:
        return 'Android'
````

````python
# find_ostype 함수에 iOS11.4.1라는 값을 넣으면 iOS라는 값을 반환합니다.
find_ostype("iOS11.4.1")
````

````python
# find_ostype 함수에 Android7.0라는 값을 넣으면 Android라는 값을 반환합니다.
find_ostype("Android7.0")
````

````python
# osversion 컬럼값을 활용해 OS의 버전을 가져오는
# find_osversion라는 이름의 함수를 정의합니다.
def find_osversion(osversion):
    # osversion에 iOS와 Android라는 값이 있으면 이를 제거합니다.
    # 이렇게 하면 최종적으로 OS의 버전만 남습니다.
    return osversion.replace("iOS", "").replace("Android", "")
````

````python
# find_osversion 함수에 iOS11.4.1이라는 값을 넣으면 11.4.1이라는 값을 반환합니다.
find_osversion("iOS11.4.1")
````

````python
# find_osversion 함수에 Android7.0이라는 값을 넣으면 7.0이라는 값을 반환합니다.
find_osversion("Android7.0")
````

````python
# 위에서 구현한 find_ostype와 find_osversion 함수에
# 판다스(Pandas)의 apply를 활용하여 모든 컬럼값을 ostype과 osversion으로 나눕니다.
# 이 결과를 ostype(clean)과 osversion(clean)이라는 이름의 새로운 컬럼으로 추가합니다.
raw_log["ostype(clean)"] = raw_log["osversion"].apply(find_ostype)
raw_log["osversion(clean)"] = raw_log["osversion"].apply(find_osversion)

# raw_log 변수에 할당된 데이터의 행렬 사이즈를 출력합니다.
# 출력은 (row, column) 으로 표시됩니다.
print(raw_log.shape)

# .head()로 raw_log 데이터의 상위 5개를 띄우되,
# osversion 컬럼과 이와 연관된 나머지 두 개의 컬럼만을 띄웁니다.
raw_log[["osversion", "ostype(clean)", "osversion(clean)"]].head()
````

````python
# osversion 컬럼값에 iOS라는 문자열이 포함되어 있는지(contains) 체크합니다.
# iOS라는 문자열이 포함되어 있다면 ostype(clean) 이라는 이름의 컬럼에 iOS라는 겂을 넣어줍니다.
raw_log.loc[raw_log["osversion"].str.contains("iOS"), "ostype(clean)"] = "iOS"

# 비슷한 방식으로, osversion 컬럼값에 Android라는 문자열이 포함되어 있다면
# ostype(clean) 컬럼에 Android라는 값을 추가합니다.
raw_log.loc[raw_log["osversion"].str.contains("Android"), "ostype(clean)"] = "Android"

# osversion 컬럼값이 iOS나 Android라는 값이 포함되어 있다면 이를 빈 값으로 바꿔줍니다. (replace)
# 이렇게 하면 osversion 컬럼에서 버전만 따로 추출할 수 있습니다.
raw_log["osversion(clean)"] = raw_log["osversion"].str.replace("iOS|Android", "")

# raw_log 변수에 할당된 데이터의 행렬 사이즈를 출력합니다.
# 출력은 (row, column) 으로 표시됩니다.
print(raw_log.shape)

# .head()로 raw_log 데이터의 상위 5개를 띄우되,
# osversion 컬럼과 이와 연관된 나머지 두 개의 컬럼만을 띄웁니다.
raw_log[["osversion", "ostype(clean)", "osversion(clean)"]].head()
````

**4. ```devicemanufacturer``` 컬럼 정보를 정리해주세요.**

```devicemanufacturer``` 컬럼은 사용자가 현재 사용하고 있는 스마트폰 디바이스를 제조한 제조사의 정보가 들어가 있습니다. 가령 ```samsung```이라고 적혀있으면 이 스마트폰은 삼성전자가 제조했다고 볼 수 있고, ```LG Electronics```라고 적혀있으면 이 스마트폰은 LG전자가 제조했다고 볼 수 있습니다. 

하지만 ```devicemanufacturer``` 컬럼에는 이 외에도 샤오미(Xiaomi), 폭스콘(Foxconn), 팬텍(PANTECH), 화웨이(HUAWEI)와 같은 마이너한 제조사 정보도 들어가 있습니다. 이 제조사들은 전체 사용자의 2%도 되지 않기 때문에, 이 데이터를 무시하거나 하나의 값(ex: Others)으로 통일해주면 데이터 분석가가 더 효율적으로 데이터를 분석할 수 있을 것 같습니다.

그러므로 ```devicemanufacturer``` 컬럼의 값을 정리해보겠습니다. 정리 방식은 다음과 같습니다.

 * ```samsung```이라고 적혀있는 값은 앞 글자를 대문자로 바꿉니다. (=```Samsung```)
 * ```LGE```와 ```LG Electronics```라는 값은 ```LG```로 통일합니다.
 * ```Apple```, ```Samsung```, ```LG```를 제외한 나머지는 ```Other```로 묶습니다.
 
최종적으로는 다음의 결과가 나와야 합니다.

  * Samsung - 230313
  * Apple - 163566
  * LG - 32649
  * Others - 7716

````python
# devicemanufacturer 컬럼값을 정리해주는
# clean_devicemanufacturer라는 이름의 함수를 정의합니다.
def clean_devicemanufacturer(devicemanufacturer):
    # devicemanufacturer 값에 samsung이라는 값이 들어가 있으면
    if "samsung" in devicemanufacturer:
        # s를 대문자(S)로 바꿔서 Samsung이라는 값을 반환해줍니다.
        return "Samsung"
    # 만일 devicemanufacturer 값에 LGE라는 값이 들어가 있으면
    elif "LGE" in devicemanufacturer:
        # LG라는 값을 반환해줍니다.
        return "LG"
    # 만일 devicemanufacturer 값에 LG Electronics라는 값이 들어가 있으면
    elif "LG Electronics" in devicemanufacturer:
        # 동일하게 LG라는 값을 반환해줍니다.
        return "LG"
    # 만일 devicemanufacturer 값에 Apple이라는 값이 들어가 있으면
    elif "Apple" in devicemanufacturer:
        # Apple이라는 값을 반환해줍니다.
        return "Apple"
    # 만일 이 모든 조건해 해당되지 않는다면
    else:
        # Other라는 값을 반환해줍니다.
        return "Other"
````

````python
# clean_devicemanufacturer 함수에 samsung이라는 값을 인자로 넣으면
# s를 대문자(S)로 바꿔서 Samsung이라는 값을 반환해줍니다.
clean_devicemanufacturer("samsung")
````

````python
# clean_devicemanufacturer 함수에 LGE라는 값을 인자로 넣으면
# LG라는 값을 반환해줍니다.
clean_devicemanufacturer("LGE")
````

````python
# clean_devicemanufacturer 함수에 LG Electronics라는 값을 인자로 넣으면
# LG라는 값을 반환해줍니다.
clean_devicemanufacturer("LG Electronics")
````

````python
# clean_devicemanufacturer 함수에 Apple이라는 값을 인자로 넣으면
# Apple이라는 값을 반환해줍니다.
clean_devicemanufacturer("Apple")
````

````python
# clean_devicemanufacturer 함수에 위의 조건에 해당하지 않는 값들을 넣어주면
# (ex: Foxconn, PANTECH, HUAWEI, BlackBerry, vivo, etc)
# Other라는 값을 반환해줍니다.
print(clean_devicemanufacturer("Foxconn"))
print(clean_devicemanufacturer("PANTECH"))
print(clean_devicemanufacturer("HUAWEI"))
````

````python
# 위에서 구현한 clean_devicemanufacturer 함수에
# 판다스(Pandas)의 apply를 활용하여 모든 devicemanufacturer 컬럼값을 정리합니다.
# 이를 devicemanufacturer(clean)이라는 이름의 새로운 컬럼으로 추가합니다.
raw_log["devicemanufacturer(clean)"] = raw_log["devicemanufacturer"].apply(clean_devicemanufacturer)

# raw_log 변수에 할당된 데이터의 행렬 사이즈를 출력합니다.
# 출력은 (row, column) 으로 표시됩니다.
print(raw_log.shape)

# devicemanufacturer(clean)에 value_counts 함수를 사용해서
# devicemanufacturer의 종류와 그 개수를 세서 출력합니다.
raw_log["devicemanufacturer(clean)"].value_counts()
````

내지는 다음과 같이 판다스(Pandas)의 [replace](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.replace.html),  [isin](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.isin.html), 그리고 [~](https://pandas.pydata.org/pandas-docs/stable/indexing.html#boolean-indexing) 기능을 활용하여 동일한 결과를 낼 수 있습니다.


````python
# 판다스의 replace 기능을 활용하여
# devicemanufacturer 컬럼값에 samsung이라는 값이 있으면
# s를 대문자(S)로 바꿔서 Samsung이라는 값으로 바꿔줍니다.
# 비슷하게 devicemanufacturer 컬럼값에 LGE, LG Electronics라는 값이 있으면
# LG라는 값으로 바꿔줍니다.
# 이를 devicemanufacturer(clean)라는 이름의 새로운 컬럼으로 바꿔줍니다.
raw_log["devicemanufacturer(clean)"] = raw_log["devicemanufacturer"].replace("samsung", "Samsung") \
                                                                    .replace("LGE", "LG") \
                                                                    .replace("LG Electronics", "LG")

# 그리고 devicemanufacturer(clean) 컬럼에 Apple, Samsung, LG라는 값이 아닌 다른 값이 있으면
# 이를 Others라는 값으로 바꿔줍니다.
raw_log.loc[~raw_log["devicemanufacturer(clean)"].isin(["Apple", "Samsung", "LG"]), "devicemanufacturer(clean)"] = "Others"

# raw_log 변수에 할당된 데이터의 행렬 사이즈를 출력합니다.
# 출력은 (row, column) 으로 표시됩니다.
print(raw_log.shape)

# devicemanufacturer(clean)에 value_counts 함수를 사용해서
# devicemanufacturer의 종류와 그 개수를 세서 출력합니다.
raw_log["devicemanufacturer(clean)"].value_counts()
````

```channel``` 컬럼을 사용자가 크몽 서비스에 유입된 경로를 기록한 정보입니다. 여기에는 웹(WEB), 구글(ex: google-play, google, google.adwords, etc), 네이버(ex: m_naver, m_naverpowercontents, pc_naver)등이 기록되어 있습니다. 이 정보도 마찬가지로 데이터를 분석하기 어렵게 산재되어 있기 때문에, 깔끔하게 정리해주고 싶습니다. 정리 방식은 다음과 같습니다.

  * 컬럼값에 google이라고 들어간 값(ex: google-play, google.adwords, etc)은 전부 google로 통일합니다.
  * 컬럼값에 daum이라고 들어간 값(ex: m_daum, etc)은 전부 daum으로 통일합니다.
  * 컬럼값이 naver라고 들어간 값(ex: m_naver, pc_naver, m_naverpowercontents, etc)은 전부 naver로 통일합니다.
  * 컬럼값에 apple이라고 들어간 값(ex: apple.searchads, etc)은 전부 apple로 통일합니다.
  * 컬럼값에 WEB이라고 들어간 값(ex: WEB)은 전부 web으로 통일합니다.
  
최종적으로는 다음의 결과가 나와야 합니다.

단, 여기서 ```channel``` 값에 빈 값(NaN, Not a Number)가 있다는 걸 주의하세요. 이 값은 소수점(float)으로 간주하기 때문에, 문자열에서 쓰이는 기능을 지원하지 않습니다. 이 점 주의해서 작성해주세요. (참고로 어떤 값이 빈 값(NaN)인지 아닌지를 구분하는 기능으로 판다스의 [isnull](https://pandas.pydata.org/pandas-docs/version/0.23.4/generated/pandas.isnull.html)을 사용하면 됩니다. 페이지의 Examples란을 살펴보세요)

````python
# 파이썬에서 수학 연산을 담당하는 numpy라는 패키지를 가져옵니다.
# 이를 np라는 축약어로 사용합니다.
import numpy as np

# 채널(channel) 컬럼에 있는 값을 정리해주는 clean_channel이라는 함수를 정의합니다.
def clean_channel(channel):
    # 만일 channel 값에 비어있는 값(NaN)이 들어가 있으면
    if pd.isnull(channel):
        # 아무런 처리도 하지 않고 그대로
        # 비어있는 값(NaN)으로 반환해줍니다.
        return np.nan

    # 만일 channel 값에 google이라는 값이 들어가 있으면(ex: google, google-play, google.adwords)
    if "google" in channel:
        # 이를 google이라는 값으로 바꿔줍니다.
        return "google"
    # 만일 channel 값에 naver라는 값이 들어가 있으면(ex: m_naver, pc_naver, m_naverpowercontents)
    elif "naver" in channel:
        # 이를 naver라는 값으로 바꿔줍니다.
        return "naver"
    # 만일 channel 값에 daum이라는 값이 들어가 있으면(ex: m_daum)
    elif "daum" in channel:
        # 이를 daum이라는 값으로 바꿔줍니다.
        return "daum"
    # 만일 channel 값에 apple이라는 값이 들어가 있으면(ex: apple.searchads)
    elif "apple" in channel:
        # 이를 apple이라는 값으로 바꿔줍니다.
        return "apple"
    # 만일 channel 값에 WEB(대문자)이라는 값이 들어가 있으면
    elif "WEB" in channel:
        # 이를 web(소문자)이라는 값으로 바꿔줍니다.
        return "web"
    # 이 모든 것에 해당하지 않으면
    else:
        # 자기 자신을 반환합니다.
        return channel
````

````python
# clean_channel 함수에 빈 값(NaN)을 인자로 넣으면
# 아무런 처리를 하지 않고 빈 값(NaN)을 반환해줍니다.
clean_channel(np.nan)
````

````python
# clean_channel 함수에 apple.searchads라는 값을 인자로 넣으면
# apple이라는 값을 반환해줍니다.
clean_channel("apple.searchads")
````

````python
# 위에서 구현한 clean_channel 함수에
# 판다스(Pandas)의 apply를 활용하여 모든 channel 컬럼값을 정리합니다.
# 이를 channel(clean)이라는 이름의 새로운 컬럼으로 추가합니다.
raw_log["channel(clean)"] = raw_log["channel"].apply(clean_channel)

# raw_log 변수에 할당된 데이터의 행렬 사이즈를 출력합니다.
# 출력은 (row, column) 으로 표시됩니다.
print(raw_log.shape)

# channel(clean)에 value_counts 함수를 사용해서
# channel의 종류와 그 개수를 세서 출력합니다.
raw_log["channel(clean)"].value_counts()
````

**6. ```inappeventcategory``` 컬럼에 들어가 있는 정보를 여러 개의 컬럼으로 나눠주세요.**

  * ```viewcategory``` - ```inappeventcategory```에서 언더바(_)를 기준으로 왼쪽 텍스트만 가져옵니다. 언더바가 없을 경우 점(.)의 왼쪽 텍스트만 가져옵니다.

  * ```viewaction``` - ```inappeventcategory```에서 점(.)의 오른쪽 텍스트만 가져옵니다.

(팁: 파이썬에서 문자열을 나누는 기능인 [split](https://wikidocs.net/13#split)을 활용하면 생각보다 쉽게 문제를 풀 수 있습니다)

````python
# inappeventcategoryd에서 viewcategory를 추출하는
# find_viewcategory라는 함수를 정의합니다.
def find_viewcategory(inappeventcategory):
    # 만일 inappeventcategory값이 비어있으면(NaN)
    if pd.isnull(inappeventcategory):
        # 아무런 처리를 하지 않고 빈 값(NaN)을 반환합니다.
        return np.nan

    # inappeventcategory에서 점(.)의 왼쪽을 가져옵니다.
    # 이 값을 chunk라는 이름의 변수에 할당합니다.
    chunk = inappeventcategory.split(".")[0]

    # 그리고 그 왼쪽에서 다시 한 번 언더바(_)의 왼쪽을 가져옵니다.
    # 이 값을 viewcategory라는 이름의 변수에 할당합니다.
    viewcategory = chunk.split("_")[0]
    
    # 이 viewcategory라는 값을 반환합니다.
    return viewcategory
````

````python
# inappeventcategoryd에서 viewid를 추출하는
# find_viewid라는 함수를 정의합니다.
def find_viewid(inappeventcategory):
    # 만일 inappeventcategory값이 비어있으면(NaN)
    if pd.isnull(inappeventcategory):
        # 아무런 처리를 하지 않고 빈 값(NaN)을 반환합니다.
        return np.nan

    # inappeventcategory에서 점(.)의 왼쪽을 가져옵니다.
    # 이 값을 viewid라는 이름의 변수에 할당합니다.
    viewid = inappeventcategory.split(".")[0]
    
    # 이 viewid라는 값을 반환합니다.
    return viewid
````

````python
# find_viewid 함수에 home.view라는 값을 인자로 넣으면
# 점(.) 기준으로 가장 왼쪽에 있는 home이라는 값을 반환합니다.
find_viewid("home.view")
````

````python
# inappeventcategoryd에서 viewaction을 추출하는
# find_viewaction이라는 함수를 정의합니다.
def find_viewaction(inappeventcategory):
    # 만일 inappeventcategory값이 비어있으면(NaN)
    if pd.isnull(inappeventcategory):
        # 아무런 처리를 하지 않고 빈 값(NaN)을 반환합니다.
        return np.nan
    
    # inappeventcategory에서 점(.)의 오른쪽을 가져옵니다.
    # 이 값을 viewaction이라는 이름의 변수에 할당합니다.
    viewaction = inappeventcategory.split(".")[1]
    
    # 이 viewaction이라는 값을 반환합니다.
    return viewaction
````

````python
# 파이썬에서 수학 연산을 담당하는 numpy라는 패키지를 가져옵니다.
# 이를 np라는 축약어로 사용합니다.
import numpy as np

# 위에서 구현한 세 개의 함수(find_viewcategory, find_viewid, find_viewaction)에
# 판다스(Pandas)의 apply를 활용하여 viewcategory, viewid, viewaction를 추출합니다.
# 이를 viewcategory, viewid, viewaction 라는 이름의 새로운 컬럼에 추가합니다.
raw_log['viewcategory'] = raw_log['inappeventcategory'].apply(find_viewcategory)
raw_log['viewid'] = raw_log['inappeventcategory'].apply(find_viewid)
raw_log['viewaction'] = raw_log['inappeventcategory'].apply(find_viewaction)

# raw_log 변수에 할당된 데이터의 행렬 사이즈를 출력합니다.
# 출력은 (row, column) 으로 표시됩니다.
print(raw_log.shape)

# .head()로 raw_log 데이터의 상위 5개를 띄우되,
# inappeventcategory 컬럼과 이와 연관된 나머지 세 개(viewcategory, viewid, viewaction)의 컬럼만을 띄웁니다.
raw_log[['inappeventcategory', 'viewcategory', 'viewid', 'viewaction']].head()
````

**8. 이름을 통일성있게 변경해주세요.**

데이터를 분석할 때 컬럼명의 형식을 동일하게 유지하는 것이 중요합니다. 가령 회사 내부 데이터 팀에서 컬럼에서 두 개의 단어를 하이픈(-)으로 연결하기로 합의했다면 모든 컬럼을 하이픈으로 연결해야 하고, 언더바(_)로 연결하기로 합의했다면 모든 컬럼을 언더바로 연결해야 합니다.

하지만 아쉽게도, 현재 데이터의 컬럼명에는 이런 통일성이 잘 갖춰져 있는 것 같습니다. (ex: ```eventcategory``` vs ```params_term```) 그러므로 데이터의 컬럼명을 작성하는 형식을 하나로 통일하도록 하겠습니다. 이번 데이터에서 컬럼명을 짓는 방식은 다음과 같습니다.

  * 단어와 단어 사이는 언제나 언더바(_)로 구분합니다. 가령 하이픈(-)으로 구분하거나, 띄어쓰기를 사용하면 안 됩니다. (ex: event_category는 허용하지만, eventcategory나 event-category, 또는 event category는 허용하지 않습니다)
  * 컬럼명에 (clean)이라고 들어간 표현은 이제는 지워줘도 될 것 같습니다. (ex channel(clean) -> channel)
