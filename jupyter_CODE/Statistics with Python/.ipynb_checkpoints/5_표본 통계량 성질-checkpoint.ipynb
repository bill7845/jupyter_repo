```python
import pandas as pd
import numpy as np 
import scipy as sp
from scipy import stats

import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline
```

```python
%precision 3
```

#### 시뮬레이션을 통한 반복 시행

```python
# 1회의 조사를 수행하는것을 시행이라고함.
# "완전히 똑같은 조건" 하에서 시행을 반복하는것도 가능


# 보통 1회의 조사로 얻을 수 있는 표본은 1개이다. 그러나 완전히 같은 조건으로(시뮬레이션)으로 반복 시행을한다면 표본이 가지는 특징을
# 명확히 알수 있게 된다.
```

```python
# 평균4, 분산0.8인 정규분포
population = stats.norm(loc=4, scale=0.8)
```

#### 표본평균을 여러 번 계산하기

```python
# 10000개의 표본평균 얻기
# 저장할 배열 생성
sample_mean_array = np.zeros(10000)
```

```python
# 배열에 표본평균 저장

np.random.seed(1) # seed(1) 고정해놓음 따라서 아래에 100000개의 sample을 seed(1)에 고정되어있음
for i in range(0,10000):
    sample = population.rvs(size=10) # 난수생성함수 rvs # 정규분포 population을 따르는 난수생성
    sample_mean_array[i] = sp.mean(sample) 
```

```python
sample_mean_array
```

```python
# 표본평균의 표준편차
sp.std(sample_mean_array, ddof=1)
```

```python
# 표본평균의 히스토그램
sns.distplot(sample_mean_array, color='black')
```

#### 샘플사이즈가 크면 표본평균은 모평균에 가까워진다

```python
# 10~100010까지 100단위로 변화하는 샘플사이즈
size_array = np.arange(start=10, stop=100100, step=100)
size_array
```

```python
# 표본평균을 저장할 변수
sample_mean_array_size = np.zeros(len(size_array))
```

```python
# 시뮬레이션 실행
np.random.seed(1)
for i in range(0,len(size_array)):
    sample = population.rvs(size=size_array[i]) # 샘플사이즈를 변화시켜가면서 반복시행
    sample_mean_array_size[i] = sp.mean(sample)
```

```python
# 샘플사이즈와 표본평균의 관계
plt.plot(size_array, sample_mean_array_size, color='black')
plt.xlabel('sample size')
plt.ylabel('sample mean')

# 샘플사이즈가 커질수록 모평균(4)에 가까워지는것을 볼 수 있음
```

#### 표본평균을 몇 번이고 계산하는 함수 만들기

```python
def calc_sample_mean(size, n_trial):
    sample_mean_array = np.zeros(n_trial)
    for i in range(0,n_trial):
        sample = population.rvs(size=size)
        sample_mean_array[i] = sp.mean(sample)
    return(sample_mean_array)
```

```python
np.random.seed(1)
sp.mean(calc_sample_mean(size=10,n_trial=10000))
```

#### 샘플사이즈를 바꿨을 때 표본평균의 분산

```python
np.random.seed(1)

#sample size 10
size_10 = calc_sample_mean(size=10, n_trial=10000)
size_10_df = pd.DataFrame({
    "sample_mean" : size_10,
    "size" : np.tile("size 10", 10000)
})

#sample size 20
size_20 = calc_sample_mean(size=20, n_trial=10000)
size_20_df = pd.DataFrame({
    "sample_mean" : size_20,
    "size" : np.tile("size 20", 10000)
})

#sample size 30
size_30 = calc_sample_mean(size=30, n_trial=10000)
size_30_df = pd.DataFrame({
    "sample_mean" : size_30,
    "size" : np.tile("size 30", 10000)
})

sim_result = pd.concat([size_10_df, size_20_df, size_30_df])

sim_result
```

```python
# 바이오플롯으로 결과확인
sns.violinplot(x='size', y='sample_mean', data=sim_result, color='gray')
```
