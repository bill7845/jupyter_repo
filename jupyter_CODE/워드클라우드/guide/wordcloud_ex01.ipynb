```python
import matplotlib.pyplot as plt  
%matplotlib inline 
from wordcloud import WordCloud, STOPWORDS

text="""
28   , 10   315.92(13.48%) . 
   159.20(19.36%) .   2098510, 
  515290 .          
2613800  . 
   2008 10 (-23.13%) 
1997 10 (-27.25%)      2008 10    .
             . 
NH    1960~2150    .          .  NH                 .
"""
spwords=set(STOPWORDS) #     
spwords.add('10') #   # ex)'should', "wouldn't", "how's", "i'm", 'then', etc ..

#   #  
wordcloud = WordCloud(max_font_size=200,font_path='c:/Windows/fonts/malgun.ttf',
                     stopwords=spwords,
                     background_color='#FFFFFF',
                     width=1200,height=800).generate(text)

# 
plt.figure(figsize=(10,8))
plt.imshow(wordcloud)
plt.tight_layout(pad=0)
plt.axis('off')

plt.savefig('wordcloud_image1.png',bbox_inches='tight') #  
plt.show()

```
