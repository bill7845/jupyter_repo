```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
%matplotlib inline
```

```python
train = pd.read_csv("C:/Users/KIHyuk/Desktop/dacon_data/santader_고객만족예측/train.csv",encoding='latin-1')
```

```python
# target val 비율

print(train.TARGET.value_counts())
print("불만족(1) 비율 : ",train[train.TARGET == 1]['TARGET'].count() / train.TARGET.count())
```

```python
train.describe()

##var3의 경우 min값이 -999999이다. NAN이나 특정 예외값을 -99999로 변환했을것이다.
```

```python
## -999999는 다른값에 비해 편차가 너무 크므로 가장 값이 많은 2로 변경
train['var3'].replace(-999999.000000,2,inplace=True)
```

```python
## 단순 식별값인 ID feature drop
train.drop('ID',axis=1,inplace=True)
```

```python
## feature , label split

x_features = train.iloc[:,:-1]
y_labels = train.iloc[:,-1]
```

```python
## train / test split
## target의 경우 불균형이므로 train과 test에 모두 비슷하게 추출됐는지 확인

from sklearn.model_selection import train_test_split

X_train,X_test, y_train,y_test = train_test_split(x_features,y_labels, test_size=0.2, random_state=0)

train_cnt = y_train.count()
test_cnt = y_test.count()
print('학습 세트 shape : ',X_train.shape,"테스트 세트 shape : ",X_test.shape)

print("학습 세트 레이블 값 분포 비율")
print(y_train.value_counts()/train_cnt)
print("\n 테스트 세트 레이블 값 분포 비율")
print(y_test.value_counts()/test_cnt)
```

## XGBoost 모델 학습과 하이퍼 파라미터 튜닝

```python
from xgboost import XGBClassifier
from sklearn.metrics import roc_auc_score

xgb_clf = XGBClassifier(n_esimators=500, random_state=156)

# 성능 평가 지표 auc, 조기 중단 파라미터100
xgb_clf.fit(X_train,y_train, early_stopping_rounds=100,
           eval_metric='auc',eval_set=[(X_train,y_train),(X_test,y_test)])

xgb_roc_score = roc_auc_score(y_test,xgb_clf.predict_proba(X_test)[:,1],average='macro')

print("ROC AUC : ",xgb_roc_score)
```

```python
## 하이퍼 파라미터 튜닝

# 차원이 매우 높으므로 과적합 가능성 높음.
# max_depth, min_child_weight, colsample_bytree 먼저 튜닝해봄.

from sklearn.model_selection import GridSearchCV

# 하이퍼 파라미터 테스트의 수행 속도를 향상시키기 위해 n_estimators를 100으로 감소
xgb_clf = XGBClassifier(n_estimators=100)

params = {'max_depth':[5,7],
         'min_child_weight':[1,3],
         'colsample_bytree':[0.5,0.75]}

# cv 3
gridcv = GridSearchCV(xgb_clf, param_grid=params, cv=3)
gridcv.fit(X_train, y_train, early_stopping_rounds=30, eval_metric='auc',eval_set=[(X_train,y_train),(X_test,y_test)])

print('GridSearchCV 최적 파라미터 : ', gridcv.best_params_)

xgb_roc_score = roc_auc_score(y_test,gridcv.predict_proba(X_test)[:,1],average='macro')
print('ROC AUC : ', xgb_roc_score)
```

```python
## 하이퍼 파라미터 튜닝 2

## 첫번째 튜닝에서 얻은 파라미터에 기반

xgb_clf = XGBClassifier(n_estimators=1000, random_state=156, learning_rate=0.02, max_depth=7, min_child_weight=1
                       ,colsample_bytree=0.75, reg_alpha=0.03)

# 성능지표를 auc, 조기 중단 파라미터값 200
xgb_clf.fit(X_train,y_train, early_stopping_rounds=200, eval_metric='auc',
           eval_set=[(X_train,y_train),(X_test,y_test)])

xgb_roc_score = roc_auc_score(y_test, xgb_clf.predict_proba(X_test)[:,1],average='macro')

print('ROC AUC : ',xgb_roc_score)
```
